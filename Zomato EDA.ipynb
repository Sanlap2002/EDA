Observation-->Maximum no of transaction are from India

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

We see the column name of the rating count is showing 0,in order to change that
we've to do this

```python
 ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

## Observation
1.We can observe that when rating is from 4.5 to 4.9 it's considered Excellent
2.We can observe that when rating is from 4.0 to 4.4 it's considered Very Good
3.We can observe that when rating is from 3.5 to 3.9 it's considered Good
4.We can observe that when rating is from 2.5 to 3.4 it's considered Average
5.We can observe that when rating is from 1.8 to 2.4 it's considered Poor


```python
ratings
```

```python

plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['black','red','orange','yellow','green','green'])
```

### Observation
 1.Not rated count very high
 2.Max no of ratings are between 2.5 and 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

The countplot only shows how many records of a specific color is present in the 
dataframe.We notice that white has only 1 record and red has 7 records
